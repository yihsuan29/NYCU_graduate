```python
def extended_euclidean(a, n):
    # Initialize values
    r1, r2 = n, a
    t1, t2 = 1, 0  # t now holds the values previously held by s
    s1, s2 = 0, 1  # s now holds the values previously held by t

    # Print the initial header
    print(f"{'t':<5} | {'s':<5} | {'r':<5} | {'q':<5}")
    print(f"{t1:<5} | {s1:<5} | {r1:<5}")
    print(f"{t2:<5} | {s2:<5} | {r2:<5}")
    
    while r2 != 0:
        q = r1 // r2  # Quotient
        r = r1 - q * r2  # Remainder
        r1, r2 = r2, r
        t = t1 - q * t2
        t1, t2 = t2, t
        s = s1 - q * s2
        s1, s2 = s2, s

        # Print s, t, r, q in the required format
        print(f"{t:<5} | {s:<5} | {r:<5} | {q:<5}")

    # At the end of the loop, r1 is the gcd and s1 is the modular inverse if r1 == 1
    if r1 == 1:
        return s1 % n  # Ensure the inverse is positive
    else:
        return None  # Inverse doesn't exist if gcd(a, n) != 1

# Test case
a = 35
n = 113
inverse = extended_euclidean(a, n)

if inverse is not None:
    print(f"The modular inverse of {a} mod {n} is: {inverse}")
else:
    print(f"No modular inverse exists for {a} mod {n}")
```

```python
import pandas as pd
import math

def square_and_multiply(base, exponent, modulus):
    """Efficiently compute base^exponent % modulus using square-and-multiply method."""
    result = 1
    base = base % modulus
    while exponent > 0:
        if exponent % 2 == 1:
            result = (result * base) % modulus
        exponent = exponent // 2
        base = (base * base) % modulus
    return result

def rabin_miller_test(n):
    """Perform the Rabin-Miller test for a given n."""
    # Express n - 1 as u * 2^t where u is odd
    u = n - 1
    t = 0
    while u % 2 == 0:
        u //= 2
        t += 1

    Z_n_star = [b for b in range(2, n) if math.gcd(b, n) == 1]  # Candidates from Z*_n
    results = []  # To store results for each b
    b_39 =[]
    for b in Z_n_star:
        bn_minus_1_mod_n = square_and_multiply(b, n - 1, n)  # Compute b^(n-1) mod n
        theta = 0  # Initialize theta(b, n)

        # Compute b^u mod n
        b_u = square_and_multiply(b, u, n)
        if b_u == 1 or b_u == n - 1:
            theta = 0  # Directly passes the condition

        # Store all b_u_2^j mod n values
        b_u_2j_values = [b_u]  # Start with b^u mod n

        # Check the sequence of powers: b^(u * 2^j) for j in range(t)
        for j in range(t):
            b_u_2j = square_and_multiply(b, u * (2**j), n)
            b_u_2j_values.append(b_u_2j)  # Store b^(u * 2^j)

            b_u_2j_plus1 = square_and_multiply(b, u * (2**(j+1)), n)
            
            if b_u_2j_plus1 == 1 and b_u_2j != 1 and b_u_2j != n - 1:
                theta = 1  # Fails primality check
                break  # No need to check further, as the test failed

        # Append the results for b, including all b_u_2j values
        results.append({
            'b': b,
            'b^(n-1) mod n': bn_minus_1_mod_n,
            'θ(b, n)': theta,
            'b_u_2j_values': b_u_2j_values
        })
        if b_u_2j_values[-1]==1 or b_u_2j_values[-1]==n-1:
            b_39.append(b)
    return results,b_39

# Example usage:
n = 39  # You can replace this with any integer n
results,b_39 = rabin_miller_test(n)

# Convert results to a DataFrame
df = pd.DataFrame(results)
print(b_39)
# Display the DataFrame (for Jupyter Notebook)
from IPython.display import display
display(df)

```

```python
import math

def square_and_multiply(base, exponent, modulus):
    """Efficiently compute base^exponent % modulus using square-and-multiply method."""
    result = 1
    base = base % modulus
    while exponent > 0:
        if exponent % 2 == 1:
            result = (result * base) % modulus
        exponent = exponent // 2
        base = (base * base) % modulus
    return result

def modular_inverse(a, m):
    """Compute the modular inverse of a mod m using the extended Euclidean algorithm."""
    m0, x0, x1 = m, 0, 1
    if m == 1:
        return 0
    while a > 1:
        q = a // m
        m, a = a % m, m
        x0, x1 = x1 - q * x0, x0
    if x1 < 0:
        x1 += m0
    return x1

def modular_sqrt(a, p):
    """Compute a^1/2 mod p using Tonelli-Shanks algorithm (if p ≡ 3 mod 4)."""
    # If p ≡ 3 (mod 4), we can use a simple method to compute square root mod p
    if p % 4 == 3:
        return square_and_multiply(a, (p + 1) // 4, p)
    else:
        raise NotImplementedError("Tonelli-Shanks algorithm for general primes is not implemented here.")

def compute_a_half_mod_pq(a, p, q):
    """Compute a^1/2 mod pq using the Chinese Remainder Theorem."""
    # Step 1: Find x1 = a^1/2 mod p
    x1 = [1, p-1]
    
    # Step 2: Find x2 = a^1/2 mod q
    x2 = [1,q-1]
    
    # Step 3: Use CRT to combine the results
    # Compute q_inverse = q^(-1) mod p and p_inverse = p^(-1) mod q
    q_inverse = modular_inverse(q, p)
    p_inverse = modular_inverse(p, q)
    
    result=[]
    
    # Now combine using the CRT formula:
    n = p * q
    for i in x2:
        for j in x1:
            r = (j * q * q_inverse + i * p * p_inverse) % n
            result.append(r)
    
    return result

# Example usage:
a = 1  # The number for which we want to compute a^1/2 mod pq
p = 5  # First prime
q = 11  # Second prime

result = compute_a_half_mod_pq(a, p, q)
print(f"a^(1/2) mod pq = {result}")
```
