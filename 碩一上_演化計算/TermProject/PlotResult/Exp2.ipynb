```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scene = ['high', 'medium', 'low']
G = [[None for _ in range(4)] for _ in range(3)]  
for i in range(3): 
    for j in range(4):  
        G[i][j] = np.load(f'./Final/crossover{j+1}_{scene[i]}.npy')
```

```python
C_index={1:'Baseline', 2:'Order Based', 3:'Priority Based', 4: "Edge Based"}
def plot(df, i,iter_):
    x = np.arange(df[0].shape[0]) 
    plt.figure(figsize=(6, 4)) 

    
    for j in range(4):
        plt.plot(x, df[j], label=f'{C_index[j+1]}')

    # Add labels and title
    plt.title(f"Performance for Scene: {scene[i]}")
    plt.xlabel('Iterations')
    plt.ylabel('Dead Nodes')
    plt.xlim(0, iter_)
    plt.legend(loc='upper right')
    plt.grid(True)
    
    plt.show()
    
X = [100,250,1000]
for i in range(3):
    plot(G[i], i,X[i])
```
